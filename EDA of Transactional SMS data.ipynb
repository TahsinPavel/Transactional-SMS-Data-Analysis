```python
#libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
df = pd.read_csv('/content/sms-data-5.csv')
df.head()
```

```python
df.describe()
```

```python
# Cleaning data with NaN value.

df = df.dropna(subset=["mobileno"])
df
```

```python
# sorting datetime

sort_requestrecievetime = df.sort_values('requestreceivetime', ascending=False)
sort_requestrecievetime
```

```python
#transaction type

t_types_df = pd.DataFrame(df.transtype.value_counts().reset_index())
t_types_df.columns = ['transtype', 'count']
t_types_df['percent'] = t_types_df['count']/t_types_df['count'].sum()
t_types_df
```

```python
# Count of Transactions by Type

sns.barplot(x="transtype", y="count", data=t_types_df)
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.barplot(x=t_types_df.transtype, y=t_types_df.percent, palette = "rocket")
ax.set_title('Transaction type distribution', fontsize = 15)
ax.set_xlabel('transaction types')
ax.set_ylabel('distribution')
plt.tight_layout()
```

```python
#Sms Status
sms_st_df = pd.DataFrame(df.smsstatus.value_counts().reset_index())
sms_st_df.columns = ['SMS_Status', 'count']
sms_st_df['percent'] = sms_st_df['count']/sms_st_df['count'].sum()
sms_st_df
```

```python
fig, ax = plt.subplots(figsize=(10,10))
sns.barplot(x=sms_st_df.SMS_Status, y=sms_st_df.percent, palette = "rocket")
ax.set_title('SMS Status distribution', fontsize = 15)
ax.set_xlabel('SMS Status')
ax.set_ylabel('distribution')
plt.tight_layout()
```

```python
# Datetime convertion


df.requestreceivetime = pd.to_datetime(df.requestreceivetime, errors="coerce", utc=True)
df.providerrequesttime = pd.to_datetime(df.providerrequesttime, errors="coerce", utc=True)
df.providerresponsetime = pd.to_datetime(df.providerresponsetime, errors="coerce", utc=True)
```

```python
# Time Difference

df['time_difference'] = (df['providerrequesttime'] - df['requestreceivetime']).dt.total_seconds()

df['time_difference']
```

```python
plt.figure(figsize=(10, 10))


plt.plot(df.index, df['time_difference'], label='Time Difference')

plt.xlabel('Index')
plt.ylabel('Time Difference (seconds)')
plt.title('Time Difference from Provider Request Time to Request Recieve Time')
plt.legend()

plt.show()
```

```python
plt.figure(figsize=(10, 10))


sns.boxplot(x= df.messageprovider, y= df.time_difference, data=df)

plt.xlabel('Message Provider')
plt.ylabel('Time Difference (seconds)')
plt.title('For Each Message Provider,Time Difference Between Request Time to Request Recieve Time')
plt.show()
```

```python
#Time Difference

df['time_difference_recieve_response'] = (df['providerresponsetime'] - df['providerrequesttime']).dt.total_seconds()

df['time_difference_recieve_response']
```

```python
# For Massage Provider


plt.figure(figsize=(10, 10))


sns.boxplot(x= df.messageprovider, y= df.time_difference_recieve_response, data=df)

plt.xlabel('Message Provider')
plt.ylabel('Time Difference (seconds)')
plt.title('For Each Message Provider,Time Difference Between Request Recieve Time to Provider Response Time')

plt.show()
```

```python
# for transaction type


plt.figure(figsize=(10, 10))


sns.boxplot(x= df.transtype, y= df.time_difference, data=df)

plt.xlabel('Message Provider')
plt.ylabel('Time Difference (seconds)')
plt.title('For Transaction Type, Time Difference Between Request Time to Request Recieve Time')

plt.show()
```

```python
plt.figure(figsize=(10, 10))


sns.boxplot(x= df.transtype, y= df.time_difference_recieve_response, data=df)

plt.xlabel('Message Provider')
plt.ylabel('Time Difference (seconds)')
plt.title('For Transaction Type,Time Difference Between Request Recieve Time to Provider Response Time')

plt.show()
```

```python
# for each sms status, trans type distribution

plt.figure(figsize=(15, 10))


sns.countplot(x= df.transtype, hue=df.smsstatus , data=df)

plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.title('Count of Each Transaction Type for Each Message Status')

plt.show()
```

```python
# for each mobile operator, sms status rate


plt.figure(figsize=(15, 10))


sns.countplot(x= df.smsstatus, hue= df.messageprovider, data=df)

plt.xlabel('SMS Status')
plt.ylabel('Count')
plt.title('SMS Status for each Massage Provider')

plt.show()
```

```python
# for each mobile operator, trans type

plt.figure(figsize=(15, 10))


sns.countplot(x= df.transtype, hue= df.messageprovider, data=df)

plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.title('Transaction Type for Each massage Provider')

plt.show()
```

```python
# create a week column to see in which day transaction request come most

weekday_dict={0:'Monday', 1:'Tuesday', 2:'Wednesday', 3:'Thursday', 4:'Friday', 5:'Saturday', 6:'Sunday'}
df['week_day'] = df['providerrequesttime'].dt.dayofweek.map(weekday_dict)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
sns.countplot(x=df.week_day, ax=ax, order=df.week_day.value_counts().index)
ax.set_title('Total transactions made on each day of the week', fontsize=18)
plt.tight_layout()
```
